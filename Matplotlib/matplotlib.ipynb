```python
import numpy as np
import matplotlib.pyplot as plt

AgeList = [10,20,30,40,50,60,60,80,80,90]
WeightList = [60,80,80,85,67,85,70,95,40,80]

AgeNP = np.array(AgeList)
WeightNP = np.array(WeightList)
```

```python
plt.plot(AgeNP,WeightNP,"r")    # plt.plot komutu ilk parametresi x  ikinci parametre y ekseni üçüncü parametreyi çizginin rengini belirler
plt.xlabel("Age")   # x eksenine başlık atar
plt.ylabel("Weight")    #y eksenine başlık atar
plt.title("Kilo Yas Grafik") # bütün grafiğe başlık atar
```

## subplot

```python
np_arr = np.linspace(0,10,20)
np_arr2 = np_arr * 3

plt.plot(np_arr,"r--") # renk kısmının yanıda çizgi yada * koyarsak grafikte verilerin geldiği yere gelirler
plt.subplot(1,2,1) #subplot fonksiyonu ile ilk parametreye grafik satırı ikincisine kolonu ve 3.ye ilk kaçıncıyı çizeceğimizi belirler
plt.plot(np_arr,np_arr2,"r--") # şuan ilk grafiği çiziyoruz
plt.subplot(1,2,2) # suan ikinci grafiği çizeceğimizi belirttik
plt.plot(np_arr2,np_arr,"g*-")
plt.show()
```

## Figure

```python
MyFigure = plt.figure() # özelleştirmek için plt sınıfından boş bir nesne türettik

figureAxes = MyFigure.add_axes([0.3,0.3,0.4,0.4]) # İLK 2 PARAMETRE BAŞLANGIÇ NOKTALARI SON 2 Sİ İSE GENİŞLİK VE YÜKSEKLİK BELİRLER
figureAxes.plot(np_arr,np_arr2,"r") 

# başlık ve x y ekseni etiketlerini belirledik
figureAxes.set_xlabel("X Ekseni")   
figureAxes.set_ylabel("Y Ekseni")
figureAxes.set_title("My Title")
```

## IC ICE GRAFİKLER

```python
MyFigure2 = plt.figure()

axes_1 = MyFigure2.add_axes([0.1,0.1,0.9,0.9])
axes_2 = MyFigure2.add_axes([0.2,0.5,0.2,0.2])

axes_1.plot(np_arr,np_arr2,"r")
axes_1.set_xlabel("X ekseni")
axes_1.set_ylabel("Y ekseni")
axes_1.set_title("big graph")

axes_2.plot(np_arr2,np_arr,"g")
axes_2.set_xlabel("X ekseni")
axes_2.set_ylabel("Y ekseni")
axes_2.set_title("small graph")
```

## subplots mantığı

```python
(myfgr , myaxs) = plt.subplots() # subplot bize tuple döndürür ilkji figüre ikincisi ise exen döndürür

myaxs.plot(np_arr,np_arr2,"r--")

(myfgr , myaxs) = plt.subplots(nrows=1,ncols=2) # subplots ta 2 kolon verirsek bize 2 kolonlu grafik verir ve eksenleri ona göre ayarlamamız gerekir

for axes in myaxs:      # 2 kolon olduğu için ve bizim eksenimiz 1 değişken oldugu için for a alıp bütün kolonları dolanıp eksenleri belirledik
   axes.plot(np_arr,np_arr2,"g")
   axes.set_xlabel("x ekseni")
   
plt.tight_layout() # bu fonksiyon iki grafik arasını açar
```

## Görsel Geliştirme

```python
new_Figure = plt.figure("""dpi=100""") # dpi özelliği çözünürlüğü ayarlamayı sağlar düştükçe kalite düşer

new_axes = new_Figure.add_axes([0.1,0.1,0.8,0.8])
new_axes.plot(np_arr,np_arr2,label="np dizi ** 2") # label özelliği hangi çizginin ne işe yaradıgını açıklar
new_axes.plot(np_arr,np_arr**3 , label = "np arr ** 3")
new_axes.legend(loc=0)   # bu fonksiyon ilede labeli gösteririz # loc özelliği bu labelin konumunu belirler


# -- Grafik Kaydetme --
#new_Figure.savefig("myfigure.png" , dpi = 200) # bu fonksiyon grafiği kaydetmemizi sağlar burdada dpi belirtebiliriz
```

## İleri Görsellik 

```python
(myfgr , myaxs) = plt.subplots()

myaxs.plot(np_arr,np_arr2,color="#C2B86C") # color diyerek rengin hex kodu ilede renk atarız
myaxs.plot(np_arr2,np_arr,"g")

plt.show()


(myfgr , myaxs) = plt.subplots()

myaxs.plot(np_arr,np_arr2 + 2,color="blue") # color diyerek rengin hex kodu ilede renk atarız
myaxs.plot(np_arr2,np_arr, color = "green" , linewidth = "8.0" , linestyle = ":") # linewidth özelliği ile çizgi kalınlığını belirleyebiliriz
myaxs.plot(np_arr2,np_arr + 8,color="red" , linewidth = "4.0" , linestyle = "-.") # linestyle özelliği ile çizgi stili belirleriz
myaxs.plot(np_arr2,np_arr + 8,color="red" , linewidth = "4.0" , linestyle = "-." , marker = "o" , markerfacecolor = "red" , markersize = 10)
#marker ile bir karakter verebiliyoruz markerfacecolor önyüz rengini markersize ilede boyut veriyoruz
```

### Stackplot

```python
o1 = [8,10,12,7,9]
o2 = [7,12,15,5,10]
o3 = [19,18,22,25,8]

years = [2011,2012,2013,2014,2015]

plt.stackplot(years,o1,o2,o3 , colors=["r","g","b"],labels=["o1","o2","o3"]) 
plt.legend()
plt.title("Yillara Göre Atılan Goller")
```

### Pie Chart(Pasta Dilimi Grafik)

```python
goals_types = ["Serbest Vurus" , "Penalti" , "Corner"]
goals = [12,3,6]
colors = ["r","g","b"]

plt.pie(goals,colors=colors,labels=goals_types,shadow=True,explode=(0.05,0.05,0.05),autopct="%1.1f%%") # pasta dilimi grafiği verir
                            # shadow gölgelendirme demek # explode ise dilimler arası aralık # autopct ise yüzdeleri belirlememizi sağlar
       
```
